```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Exponencial (x,lambda_=1):
    return lambda_ * np.exp( -lambda_ * x)
```

```python
Exponencial (0, lambda_=2)
```

linespace

```python
x = np.linspace (0.,3.,11)
```

```python
x
```

```python
y1 = Exponencial (x, lambda_=0.5)
y2 = Exponencial (x, lambda_=1.)
y3 = Exponencial (x, lambda_=1.5)
```

```python
y1
y2
y3
```

```python
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.scatter(x, y3)
```

```python
plt.plot(x, y1, label=r'$\lambda= 0.5$')
plt.plot(x, y2, label=r'$\lambda= 1.$')
plt.plot(x, y3, label=r'$\lambda= 1.5$')
plt.xlabel(r'$x$', fontsize = 12)
plt.ylabel(r'$P(x)$', fontsize = 12)
plt.xlim(0.,2.)
plt.legend(fontsize=12)
```
