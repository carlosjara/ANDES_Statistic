```python
Data = pd.read_csv('Database/Ventas.csv')
```

```python
# Tomamos las filas de los datos
Columns = np.array(Data.iloc[:,1:4])
```

```python
A = np.zeros(3)
```

```python
Percentiles = np.zeros((3,5))
```

```python
for i in range(3):
    Percentiles[i] = np.percentile( Columns[:,i], [0,25,50,75,100] )
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)

ax.set_xticklabels(['2018','2019','2020'],rotation=20, fontsize=20)

for i in range(3): # Los años de los datos
    for j in range(5): # Los cuartiles
        ax.text( i+1.2, Percentiles[i][j],Percentiles[i][j],color='r' )


ax.boxplot(Columns)
```
