```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
def Likelihood(x, mu, sigma=30): #Verosimiltud
    return norm.pdf(x, loc=mu, scale=sigma)
```

```python
mu = np.linspace(-50,60,100)
y = Likelihood(mu,0)
```

```python
plt.plot(mu,y)
```

```python
# ventas diarias
x = np.array([1000,1200,1400])
```

```python
# Modelos aproximados (tecnica)
N = 20 # Numero de puntos
p = 1/N
mu = np.arange(0,2000+p,p)
len(mu)
```

```python
mu[10000]
```

```python
# definir la informacion prior (no informativa cualquier venta es posible)
Prior = np.ones_like(mu)
Prior[:] = p
```

```python
plt.plot(mu,Prior)
```

```python
Likelihood = norm.pdf( mu, loc = x[0], scale=50)
```

```python
plt.plot(mu, Likelihood) 
```

```python
# leyendo la formula . la probabilidad de que se vendieron mil por la probabilidad de venta de cada uno
plt.plot(mu, Likelihood*Prior)
```

```python
Prob_x = np.sum( Likelihood*Prior )
```

```python
Prob_x
```

```python
# Posterior (probabilidad de observar mil dado tal modelo)
Posterior = Likelihood*Prior/Prob_x
```

```python
plt.plot(mu, Posterior)
```

```python
# dia 2
# Posterior se convierte en la Prior
Prior = Posterior
Posterior = Likelihood*Prior/np.sum( Likelihood*Prior )
```

```python
Likelihood = norm.pdf( mu , loc=x[1], scale = 50)
```

```python
plt.plot(mu, Prior)
#plt.plot(mu, Likelihood)
plt.plot(mu, Likelihood*Prior/np.sum( Likelihood*Prior ), color ='r')
```

```python
mean = np.sum(mu*Posterior) # frecuencia relativa por los valores de cada categoria
```

```python
mean
```

```python
# dia 3
# Posterior se convierte en la Prior
Prior = Posterior
Prob_x = np.sum( Likelihood*Prior )
Posterior = Likelihood*Prior/np.sum( Likelihood*Prior )
# actualizamos con venta real del dia 3 con la pos 2 de array
Likelihood = norm.pdf( mu , loc=x[2], scale = 50)
plt.plot(mu, Prior)
#plt.plot(mu, Likelihood)
plt.plot(mu, Likelihood*Prior/np.sum( Likelihood*Prior ), color ='r')
```

```python
mean = np.sum(mu*Posterior) # frecuencia relativa por los valores de cada categoria
```

```python
# ventas diarias
x = np.array([1000,1200,1400,1300,980,1000,1500,1010])
Prediccion = np.zeros_like (x)
Prior = np.ones_like(mu)
Prior[:] = p
for i in range (len(x)):
    Likelihood = norm.pdf( mu , loc=x[i], scale = 50)
    Prob_x = np.sum( Likelihood*Prior)
    
    Posterior = Likelihood*Prior/np.sum( Likelihood*Prior )
    
    if i != 1:
        plt.plot(mu,Posterior)
    mean = np.sum(mu*Posterior)
    Prediccion[i] = mean
    sigma2 = np.sum(Posterior*(mu-mean)**2)
    print(mean,np.sqrt(sigma2))
    Prior = Posterior
```

```python
plt.plot(x, label='Observaciones')
plt.plot(Prediccion, label='Prediccion')
plt.legend()
```
