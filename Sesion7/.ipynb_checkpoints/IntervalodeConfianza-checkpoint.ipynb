```python
#Caso pasticula
s = np.array([95,108,97,112,99,106,105,100,99,98,104,110,107,111,103,110])
s.shape
```

```python
# parametro de distribucion dado los datos entregado
#calculamos la media muestral
mean = np.mean(s)
x = np.linspace(90,120,50)
std = np.std(s)
#std = std/np.sqrt(len(s))
y = norm.pdf(x, loc=mean, scale=std)
plt.hist(s,density=True)
plt.plot(x,y)
```

```python
M = norm.rvs(size=10000, loc=mean, scale=std)
```

```python
I1 = np.percentile(M,2.5)
I2 = np.percentile(M,97.5)
```

```python
print(I1,I2)
```

Un fabricante de electrodomesticos sabe que la vida media de estos, sigue una dist normal con media 100 meses y desviacion tipica de 12 meses
1. Cual es el valor limiete de tiempo de vida para no hacer reclamo al fabricante
2. Cuales la probabilidad de que el articulo dure mas de 140 meses


```python
ValorCritico = norm.ppf(0.05, loc=100, scale=12)
x = np.linspace(50,200)
y = norm.pdf(x , loc=100,scale=12)
plt.plot(x,y)
plt.axvline(ValorCritico, color='r')
```
