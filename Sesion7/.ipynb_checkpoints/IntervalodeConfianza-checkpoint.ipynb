```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x= np.linspace(-5,5,1000)
y = norm.pdf(x , loc=0, scale=1) #media = Loc, Desviacion estandar = scale
```

```python
M = norm.rvs(size=10000, loc=0, scale=1)
```

```python
Q1= np.percentile(M,25)
Q2= np.percentile(M,50)
Q3= np.percentile(M,75)
```

```python
print("El 50% de la informacion esta entre : ",Q1," - ",Q3)
```

```python
plt.plot(x,y)
plt.hist(M, density=True, bins=50)
plt.axvline(x=Q1)
plt.axvline(x=Q2, color='r')
plt.axvline(x=Q3)
```

```python
plt.boxplot(M)
```

```python
#Caso particular
s = np.array([95,108,97,112,99,106,105,100,99,98,104,110,107,111,103,110])
s.shape
```

```python
# parametro de distribucion dado los datos entregado
#calculamos la media muestral
mean = np.mean(s)
x = np.linspace(90,120,50)
std = np.std(s) # De datos NO AGRUPADOS
#std = std/np.sqrt(len(s))
y = norm.pdf(x, loc=mean, scale=std)
plt.hist(s,density=True)
plt.plot(x,y)
```

```python
M = norm.rvs(size=10000, loc=mean, scale=std)
```

```python
I1 = np.percentile(M,2.5)
I2 = np.percentile(M,97.5)
```

```python
print(I1,I2)
print("El 50% de la informacion esta entre : ",I1," - ",I2)
print("Con un 95% confianza el valor esta entre : ",I1," - ",I2)
```

Un fabricante de electrodomesticos sabe que la vida media de estos, sigue una dist normal con media 100 meses y desviacion tipica de 12 meses
1. Cual es el valor limite de tiempo de vida para no hacer reclamo al fabricante
2. Cuales la probabilidad de que el articulo dure mas de 140 meses

Esto es prueba de una cola por que son temas de produccion


```python
ValorCritico = norm.ppf(0.05, loc=100, scale=12)
x = np.linspace(50,200)
y = norm.pdf(x , loc=100,scale=12)
plt.plot(x,y)
plt.axvline(ValorCritico, color='r')
```

```python
ValorCritico # Respuesta 1
#Significancia ...
#los valores con menor que 80 (que es el 5% ) 
#podria decirse que la vida media es menor a 100
# con eso refuto la hipotesis de la vida media es 100
```

```python
norm.cdf(140, loc=100, scale=12)
```

```python
(1-norm.cdf(140, loc=100, scale=12))*100 # Respuesta 2
```
